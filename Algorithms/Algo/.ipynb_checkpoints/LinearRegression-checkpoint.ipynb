<h3 style="text-align:center;"><u>SIMPLE LINEAR REGRESSION</u></h3>

**Dataset :**

Assume that you have recorded the diameters and prices of pizza that you have previously eaten.
<br><br>
<br>

**Training instance**      |    **Diameter (in inches)** | **Price(in dillars)**
---------------------------|-------------------------------|----------------------------
1.                         |            6                  |           7
2.                         |            8                  |           9
3.                         |            10                 |           13
4.                         |            14                 |           17.5
5.                         |            18                 |           18

<br>

<h5>Plot the graph for the dataset : </h5>

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
diameter = np.array([6,8,10,14])
price = np.array([7,9,13,17.5])

#plot
plt.plot(diameter,price,'r.',label="data point",markersize=15)
plt.title("Relationship between diameter and price")
plt.xlabel("$Diameter(in)$")
plt.ylabel("Price")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#initialize random w and b

w = 0
b = 0

#define a function for prediction

def predict(x):
    return (w*x + b)


#define a function to calculate the gradient Loss wrt w

def gradientw(x,y,w,b):
    return x*((w*x+b)-y)

#define a function to calculate gradient wrt b

def gradientb(x,y,w,b):
    return ((w*x+b)-y)

#define a function to calculate loss

def loss(x,y):
    y_pred=predict(x)
    return ((y_pred-y)*(y_pred-y))/2

#define a step length for gradient descent(leaning rate)

lr=0.1

#create three lists to store weight and bias and loss 

weight_list = []

bias_list = []

loss_list = []

#prediction before training 

print("18 diameter pizza costs : ",predict(18)," dollars")

for epoch in np.arange(100):
    
    print("Epoch number : ",epoch)
    
    loss_sum = 0
    
    for x,y in zip(diameter,price):
        
        gradw = gradientw(x,y,w,b)
        
        gradb = gradientb(x,y,w,b)
        
        loss_value = loss(x,y)
        
        w = w-lr*gradw
        
        b = b-lr*gradb
        
        loss_sum+=loss_value
        
        loss_list.append(loss_sum)
        
        weight_list.append(w)
        
        bias_list.append(b)
        
        print("Weight ; ",w)
        
        print("Bias : ",b)
        
        print("Gradient wrt Weight : ",gradw)
        
        print("Gradient wrt bias : ",gradb)
        
    print("Mean Error : ",loss_sum/diameter.size)    
                       
print("18 diameter pizza costs : ",predict(18)," dollars")
       
```
