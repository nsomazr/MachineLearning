```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

#create an input data and their corresponding labels

input_data=np.array([1,2,3,4,5])

output_data=np.array([2,4,6,8,10])

#initialize random w and b

w = 0
b = 0

#define a function for the forward path

def forward(x):
    return (w*x + b)


#define a function to calculate the gradient Loss wrt w

def gradientw(x,y,w,b):
    return x*((w*x+b)-y)

#define a function to calculate gradient wrt b

def gradientb(x,y,w,b):
    return ((w*x+b)-y)

#define a function to calculate loss

def loss(x,y):
    y_pred=forward(x)
    return ((y_pred-y)*(y_pred-y))/2

#define a step length for gradient descent(leaning rate)

lr=0.1

#create three lists to store weight and bias and loss 

weight_list = []

bias_list = []

loss_list = []

#prediction before training 

print("7 hours of study result to ",forward(7)," points")

for epoch in np.arange(100):
    
    print("Epoch number : ",epoch)
    
    loss_sum = 0
    
    for x,y in zip(input_data,output_data):
        
        gradw = gradientw(x,y,w,b)
        
        gradb = gradientb(x,y,w,b)
        
        loss_value = loss(x,y)
        
        w = w-lr*gradw
        
        b = b-lr*gradb
        
        loss_sum+=loss_value
        
        loss_list.append((loss_sum/input_data.size))
        
        weight_list.append(w)
        
        bias_list.append(b)
        
        print("Weight ; ",w)
        
        print("Bias : ",b)
        
        print("Gradient wrt Weight : ",gradw)
        
        print("Gradient wrt bias : ",gradb)
                       
print("7 hours of study result to {:.1f}".format(forward(7))," points")        
        
        
        
        
```
