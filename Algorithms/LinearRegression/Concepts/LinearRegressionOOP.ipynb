```python
import matplotlib as mtl
from matplotlib import pyplot as plt
from math import pow
import numpy as np
```

```python
diameter = np.array([6,8,10,14,18])
price = np.array([7,9,13,17.5,18])
```

```python
class LinearRegression():
    def __init__(self,w=0,b=0):
        self.w = w
        self.b = b
    def predict(self,x):
        return (w*x + b)   
    def gradients(self,x,y):
        gradw = x*(predict(x)-y)
        gradb = (predict(x)-y)
        return gradw,gradb
    def loss(self,x,y):
        return (pow((predict(x)-y),2))/2
    def fit(self,x,y,lr=0.001):
        for epoch in np.arange(1,2001):
            #initialize the sum of loss
            loss_sum = 0
            #for each epoch pass through all datapoints
            for x,y in zip(x,y):
                gradw,gradb = gradients(x,y)
               #update weight
                w = w - lr*gradw
                #update bias
                b = b- lr*gradb
                #sum loss
                loss_sum += loss(x,y)
```

```python
model = LinearRegression()
```

```python
model.fit(diameter,price)
```
